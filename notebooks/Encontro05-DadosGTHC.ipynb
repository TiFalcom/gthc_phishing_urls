# Cria√ß√£o de Vari√°veis

```python
#Processo que envolve criatividade e t√©cnica
```

## Vari√°veis calculaveis

```python
#Percentual de Pontos na minha URL
df['perc_pontos'] = df['n_dots'] / df['url_length']
```

```python
#Percentual de Hifens na minha URL
df['perc_hifens'] = df['n_hypens'] / df['url_length']
```

```python
#Percentual de Hifens e Pontos na minha URL
df['perc_hifens_pontos'] = (df['n_hypens'] + df['n_dots']) / df['url_length']
```

```python
df.describe()
```

## Vari√°veis Discretizadas

```python
#Porte da URL
def funcao_porte_url(x):
    if x < 50:
        return 'PEQUENA'
    if x < 100:
        return 'MEDIA'
    return 'GRANDE'

df['porte_url'] = df['url_length'].apply(funcao_porte_url)
```

```python
df[['url_length', 'porte_url']]
```

```python
#Tem ponto de interroga√ß√£o?
def funcao_ponto_interrogacao(x):
    if x > 0:
        return 'SIM'
    return 'NAO'

df['tem_ponto_interrogacao'] = df['n_questionmark'].apply(funcao_ponto_interrogacao)
```
