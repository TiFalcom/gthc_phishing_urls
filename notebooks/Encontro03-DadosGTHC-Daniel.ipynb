```python
#Criando conexão
conn = psycopg2.connect(
    host=db_params['host'],
    database=db_params['database'],
    user=db_params['user'],
    password=db_params['password']
)
```

# Criação de Tabela

```python
# Create a cursor object
cur = conn.cursor()

# Set automatic commit to be true, so that each action is committed without having to call conn.committ() after each command
conn.set_session(autocommit=True)
```

```python
# Criar tabela
cur.execute('''
    CREATE TABLE teste_daniel (
    age INT,
    gender VARCHAR(10),
    income DECIMAL(10, 2),
    education VARCHAR(50),
    marital_status VARCHAR(20),
    number_of_children INT,
    home_ownership VARCHAR(20),
    credit_score VARCHAR(20)
);

''')
```

```python
#Fechando conexão, boa prática
cur.close()
conn.close()
```

# Inserção de Dados

```python
#Definindo a engine do banco de dados
engine = create_engine(f'postgresql://{db_params["user"]}:{db_params["password"]}@{db_params["host"]}/{db_params["database"]}')
```

```python
# Lendo os dados
df = pd.read_csv(os.path.join(project_dir, 'data', 'raw', 'Credit Score Classification DatasetValidated.csv')).drop(columns= 'Unnamed: 0')
df.head()
```

```python
df.to_sql('teste_daniel', engine, if_exists='replace', index=False)
```
